```python
import pandas as pd
import numpy as np

# Load your dataset
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Salaries.csv')
df.head()
```

### **Basic Data Exploration**

```python
# Display the number of rows and columns
print("Number of rows: ", df.shape[0])
print("Number of columns: ", df.shape[1])

# Display data types of each column
print("\nData types of each column:")
print(df.dtypes)

# Check for missing values in each column
print("\nMissing values in each column:")
print(df.isnull().sum())
```

```python
df.info()
```

## **Descriptive Statistics**

```python
# Basic statistics
mean_salary = df['TotalPay'].mean()
median_salary = df['TotalPay'].median()
mode_salary = df['TotalPay'].mode()[0]  # Mode can have multiple values, so we take the first one
min_salary = df['TotalPay'].min()
max_salary = df['TotalPay'].max()
salary_range = max_salary - min_salary
std_dev_salary = df['TotalPay'].std()

# Display the results
print(f"Mean Salary: {mean_salary:.2f}")
print(f"Median Salary: {median_salary:.2f}")
print(f"Mode Salary: {mode_salary:.2f}")
print(f"Minimum Salary: {min_salary:.2f}")
print(f"Maximum Salary: {max_salary:.2f}")
print(f"Salary Range: {salary_range:.2f}")
print(f"Standard Deviation of Salary: {std_dev_salary:.2f}")
```

```python
df['TotalPay'].describe()
```

## **Data Cleaning**

```python
# Remove rows with missing values
df_cleaned = df.dropna()
```

```python
df.isnull().sum()
```

```python
# Impute with mean for BasePay
df['BasePay'].fillna(df['BasePay'].mean(), inplace=True)

# Impute with mean for OvertimePay and OtherPay
df['OvertimePay'].fillna(df['OvertimePay'].mean(), inplace=True)
df['OtherPay'].fillna(df['OtherPay'].mean(), inplace=True)
```

```python
# 'Notes' is not critical for your analysis, consider dropping the column.
# Drop 'Notes' column
df.drop('Notes', axis=1, inplace=True)
```

```python
# Impute with mean for Benefits
df['Benefits'].fillna(df['Benefits'].mean(), inplace=True)
```

```python
#Status' is not critical for your analysis
# Drop 'Status' column
df.drop('Status', axis=1, inplace=True)
```

## Basic Data Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Distribution of Salaries - Histogram
plt.figure(figsize=(10, 6))
sns.histplot(df['TotalPay'], bins=30,  color='skyblue')
plt.title('Distribution of Salaries')
plt.xlabel('Total Pay')
plt.ylabel('Frequency')
plt.show()
```

```python
# Proportion of Employees in Different Departments - Pie Chart
plt.figure(figsize=(10, 6))
department_counts = df['JobTitle'].value_counts().head(15)  # Assuming you want to show the top 15 departments
plt.pie(department_counts, labels=department_counts.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
plt.title('Proportion of Employees in Different Departments')
plt.show()
```

## Grouped Analysis

```python
# Assuming 'JobTitle' is the column by which you want to group the data
grouped_df = df.groupby('JobTitle') ['TotalPay'].agg(['count', 'mean', 'median', 'min', 'max', 'std']).reset_index()

# Display the summary statistics
print(grouped_df.head().to_markdown(index=False))
```

```python
# Compare average salaries across different groups
average_salaries = grouped_df.sort_values(by='mean', ascending=False)
print(average_salaries.head().to_markdown(index=False))
```

## Simple Correlation Analysis

```python
# Scatter plot to visualize the relationship
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['TotalPay'], y=df['BasePay'], color='blue', alpha=0.7)
plt.title(f'Scatter Plot: TotalPay vs BasePay')
plt.xlabel('TotalPay')
plt.ylabel('BasePay')
plt.show()
```

```python
# Calculate the correlation matrix
correlation_matrix = df.corr()

# Extract the correlation with 'BasePay' column
correlation_with_basepay = correlation_matrix['BasePay']

# Plotting a heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```
